```python
df.Ship_Mode.unique()
```

**DATA ANALYSIS OF SHIP MODE**

```python
plt.figure(figsize=(5,5))
df1=sns.countplot(data=df, x='Ship_Mode', hue='Ship_Mode')
df1.bar_label(df1.containers[0])
df1.bar_label(df1.containers[1])
plt.show()
```

```python

df1=df.groupby(['Ship_Mode'], as_index=False).sum().sort_values(by='Profit', ascending=False)
plt.figure(figsize=(5,5))
df1=sns.barplot(data=df1, x='Ship_Mode',y='Profit', hue='Ship_Mode' )
df1.bar_label(df1.containers[0])
df1.bar_label(df1.containers[1])
plt.show()
```

*The visual representation illustrates that the `Ground Level (shipping mode)` has yielded higher profits compared to the ` Air level (shipping mode)`.*

**DATA ANALYSIS OF REGION**

```python
plt.figure(figsize=(5,5))
df1=sns.countplot(data=df, x='Region',order=df["Region"].value_counts().index, palette='Set1')
df1.bar_label(df1.containers[1])
df1.bar_label(df1.containers[2])
df1.bar_label(df1.containers[3])
df1.bar_label(df1.containers[0])
plt.show()
```

```python
df1=df.groupby(['Region'], as_index=False)['Profit'].sum().sort_values(by="Profit", ascending=False)
plt.figure(figsize=(5,5))
regions = df1['Region']
profits = df1['Profit']
plt.pie(profits, labels=regions, autopct="%1.2f%%", startangle=90)
plt.title('Profit Distribution by Region')
plt.show()
```

*The graphical representation indicates that the profits generated in the `South Region` surpass those of the remaining `regions`.*

**DATA ANALYSIS OF CATEGORY**

```python
plt.figure(figsize=(5,5))
df1=sns.countplot(data=df, x='Category', hue='Category')
df1.bar_label(df1.containers[0])
df1.bar_label(df1.containers[1])
df1.bar_label(df1.containers[2])
plt.show()
```

```python
df1=df.groupby(['Category'], as_index=False)['Profit'].sum()
plt.figure(figsize=(5,5))
df1=sns.stripplot(data=df1, x='Category',y='Profit',hue='Category')
plt.show()
```


*The graphs depict a higher profit margin from the `'Furniture and Electronics'` category in comparison to the profits generated by `'Office Supplies'`.*

**DATA ANALYSIS OF SUB_CATEGORY**

```python
plt.figure(figsize=(10,10))
df1=sns.countplot(data=df, x='Sub_Category', hue='Sub_Category')
for bars in df1.containers:
    df1.bar_label(bars)
plt.xticks(rotation=90, ha='right') 
plt.show()
```

```python
df1=df.groupby(['Sub_Category'], as_index=False)['Profit'].sum()
plt.figure(figsize=(10,10))
df1=sns.scatterplot(data=df1, x='Sub_Category',y='Profit', hue='Sub_Category')
plt.xticks(rotation=90, ha='right') 
plt.show()
```


*The profit analysis from the visual data suggests that `'Laptops'` outperformed other items in terms of generating revenue.*

**DATA ANALYSIS OF STATE**

```python
plt.figure(figsize=(5,5))
df1=sns.countplot(data=df, x='State',hue='State')
for bars in df1.containers:
    df1.bar_label(bars)
plt.show()
```

```python
df1=df.groupby(['State'], as_index=False)['Profit'].sum()
plt.figure(figsize=(10,10))
df1=sns.catplot(data=df1, x='State',y='Profit',hue='State')
plt.show()
```

*The graph highlights that doing business in 'Texas and Illinois' is more lucrative than in other states.*

**DATA ANALYSIS OF SUb_CATEGORY**

```python
plt.figure(figsize=(10,10))
df1=sns.countplot(data=df, x='Sales', hue='Sales')
for bars in df1.containers:
    df1.bar_label(bars)
plt.show()
```

```python
df1=df.groupby(['Sub_Category'], as_index=False)['Sales'].sum().sort_values(by='Sub_Category')
plt.figure(figsize=(10,10))
df1=sns.barplot(data=df1, x='Sub_Category',y='Sales', hue='Sub_Category')
 
plt.xticks(rotation=90, ha='right') 
plt.show()
```

*from the above grapgh we concluded that `laptops` sold more than other things*
